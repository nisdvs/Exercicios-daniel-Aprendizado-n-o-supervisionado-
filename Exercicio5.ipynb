```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dados_maquinas = pd.DataFrame({
    'Máquina': ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10'],
    'Temperatura (°C)': [70.2, 65.1, 75.5, 80.3, 68.7, 72.9, 78.6, 66.4, 73.1, 69.5],
    'Vibração': [12.5, 8.2, 15.6, 10.2, 11.8, 14.3, 9.8, 8.9, 13.7, 12.1],
    'Corrente': [4.7, 3.9, 5.1, 4.5, 4.2, 5.3, 4.8, 4.0, 5.0, 4.3]
})
```

```python
dados_maquinas = dados_maquinas.drop('Máquina', axis=1)
```

```python
kmeans = KMeans(n_clusters=3)
dados_maquinas['Cluster'] = kmeans.fit_predict(dados_maquinas)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([
    dados_maquinas[dados_maquinas['Cluster'] == 0]['Temperatura (°C)'],
    dados_maquinas[dados_maquinas['Cluster'] == 1]['Temperatura (°C)'],
    dados_maquinas[dados_maquinas['Cluster'] == 2]['Temperatura (°C)'],
],
    labels=['Grupo 1', 'Grupo 2', 'Grupo 3']
)
plt.title('Agrupamento de Máquinas por Temperatura (°C)')
plt.xlabel('Grupos')
plt.ylabel('Temperatura (°C)')
plt.show()
```
