```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dados_setores = pd.DataFrame({
    'Setores': [1, 2, 3, 4, 5, 6, 7, 8],
    'Número de Produtos Fabricados': [100, 50, 15, 200, 500, 1000, 375, 450]
})
```

```python
kmeans = KMeans(n_clusters=2)  # Definindo 2 clusters
kmeans.fit(dados_setores[['Número de Produtos Fabricados']])
```

```python
sse = kmeans.inertia_
print("SSE:", sse)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dados_setores['Setores'], dados_setores['Número de Produtos Fabricados'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 0], marker='x', color='red')
plt.xlabel('Setores')
plt.ylabel('Número de Produtos Fabricados')
plt.title('Segmentação de Setores por Produtividade')
plt.show()
```

```python
dados_setores['Grupo'] = labels
grouped_data = dados_setores.groupby('Grupo').mean()
print(grouped_data)
```
