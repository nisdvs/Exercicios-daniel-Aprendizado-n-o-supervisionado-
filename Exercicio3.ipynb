```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dados_dias_semana = pd.DataFrame({
    'Dia da Semana': [1, 2, 3, 4, 5, 6, 7],
    'Número de Clientes': [15, 20, 30, 100, 350, 500, 700]
})
```

```python
kmeans = KMeans(n_clusters=2)  # Definindo 2 clusters
kmeans.fit(dados_dias_semana[['Número de Clientes']])
```

```python
sse = kmeans.inertia_
print("SSE:", sse)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dados_dias_semana['Dia da Semana'], dados_dias_semana['Número de Clientes'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 0], marker='x', color='red')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Clientes')
plt.title('Agrupamento de Dias de Semana por Movimento de Clientes')
plt.show()
```

```python
dados_dias_semana['Grupo'] = labels
grouped_data = dados_dias_semana.groupby('Grupo').mean()
print(grouped_data)
```
